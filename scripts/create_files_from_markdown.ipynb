```python



image_template = '''			<div class="species-gallery">
				<div>
					<img src="img/noca2.jpg" alt="" />
					<p class="caption">Female; Photo by Marie Read</p>
				</div>
				<div>
					<img src="img/noca1.jpg" alt="" />
					<p class="caption">Male; Photo by Brian E. Small</p>
				</div>
				<div>
					<img src="img/noca2.jpg" alt="" />
					<p class="caption">Female; Photo by Marie Read</p>
				</div>
				
				
			</div>
'''

an_image = '''				<div>
					<img src="photos/{photo_file}.jpg" alt="" />
					<p class="caption">{caption}</p>
				</div>'''



def get_actual_image_name(image_file, actual_images):
    if image_file in actual_images:
        return image_file
    else:
        distance = 10
        best = "ERROR"
        for im in actual_images:
            d = editdistance.eval(image_file, im)
            if d < distance:
                best = im
                distance = d 
        print(best, image_file)
        return best

mypath = "../photos/"
actual_images = [f for f in listdir(mypath) if isfile(join(mypath, f))]
actual_images = map(lambda x: x.replace(".jpg",""), actual_images)

for i in range(1,df_pages['Sequence of all items'].max()):
    index = df[df['Species file name']==df_pages.SpeciesFile[i]].index[0]
    d = df.iloc[index]
    
    image_table_text = ''
    if type(d["Photo #1"]) == unicode:
        image_table_text += '<div class="species-gallery">'
        photo_file = d["Photo #1"].replace(" ","_").replace("'","")    
        photo_file = get_actual_image_name(photo_file, actual_images)
        
        
        caption = d["Caption for Photo #1"]
        
        image_table_text += an_image.format(photo_file = photo_file ,
                                           caption = caption )
        
        if type(d["Photo #2"]) == unicode:
            photo_file = d["Photo #2"].replace(" ","_").replace("'","")
            photo_file = get_actual_image_name(photo_file, actual_images)
      
            
            caption = d["Caption for Photo #2"]    
            image_table_text += an_image.format(photo_file = photo_file ,
                                           caption = caption )
            
        if type(d["Photo #3"]) == unicode:
            photo_file = d["Photo #3"].replace(" ","_").replace("'","")
            photo_file = get_actual_image_name(photo_file, actual_images)

            caption = d["Caption for photo #3"]    
            image_table_text += an_image.format(photo_file = photo_file ,
                                           caption = caption )
        image_table_text += "</div>"
        
        
    with open("Species_Image_HTML/"+d['Species file name']+"_images.html", 'w') as f:
        f.write(image_table_text)


    
```

```python
def get_group_name(species_name):
    try:
        value = df_species[df_species.Species == species_name]['Group name'].values[0]
        return value
    except:
        best = 'error'
        distance = 30
        for index, row in df_species.iterrows():
            d = editdistance.eval(row.Species, species_name)
            if d < distance:
                best = row['Group name']
                distance = d 
        return best
    
def get_species_name(species_name):
    try:
        value = df_species[df_species.Species == species_name]['Species'].values[0]
        return value
    except:
        best = 'error'
        distance = 30
        for index, row in df_species.iterrows():
            d = editdistance.eval(row.Species, species_name)
            if d < distance:
                best = row['Species']
                distance = d 
        return best

def get_family_name(species_name):
    try:
        value = df_species[df_species.Species == species_name]['Family name'].values[0]
        return value
    except:        
        best = 'error'
        distance = 30
        for index, row in df_species.iterrows():
            d = editdistance.eval(row.Species, species_name)
            if d < distance:
                best = row['Family name']
                distance = d 
        return best


df['group_name'] = df['Species name'].apply(get_group_name)

df['family_name'] = df['Species name'].apply(get_family_name)

df['species_name2'] = df['Species name'].apply(get_species_name)

```
