```python
# note that the page values start at 1, so add a value to the start of these
# arrays
species_files = ['0']
species_names_headers = ['0']
web_chapters = ['0']
subchapters = ['0']
from_pages = ['0']
see_alsos = ['0']


for i in range(1,df_pages['Sequence of all items'].max()):
    index = df[df['Species file name']==df_pages.SpeciesFile[i]].index[0]
    d = df.iloc[index]
    species_names_headers.append(d['Species name'])
    species_files.append(d['Species file name'])
    web_chapters.append(d['Web chapter heading'].replace(".",":"))
    subchapters.append(d['Web subchapter heading'].replace('"',"&quot;").replace('--','&mdash;'))
    from_pages.append(d['Book page'])
    see_alsos.append('')
    
    
    

    
```

```python



image_template = '''			<div class="species-gallery">
				<div>
					<img src="img/noca2.jpg" alt="" />
					<p class="caption">Female; Photo by Marie Read</p>
				</div>
				<div>
					<img src="img/noca1.jpg" alt="" />
					<p class="caption">Male; Photo by Brian E. Small</p>
				</div>
				<div>
					<img src="img/noca2.jpg" alt="" />
					<p class="caption">Female; Photo by Marie Read</p>
				</div>
				
				
			</div>
'''

an_image = '''				<div>
					<img src="photos/{photo_file}.jpg" alt="" />
					<p class="caption">{caption}</p>
				</div>'''


for i in range(1,df_pages['Sequence of all items'].max()):
    index = df[df['Species file name']==df_pages.SpeciesFile[i]].index[0]
    d = df.iloc[index]
    
    image_table_text = ''
    if type(d["Photo #1"]) == unicode:
        image_table_text += '<div class="species-gallery">'
        photo_file = d["Photo #1"].replace(" ","_").replace("'","")
        caption = d["Caption for Photo #1"]
        
        image_table_text += an_image.format(photo_file = photo_file ,
                                           caption = caption )
        
        if type(d["Photo #2"]) == unicode:
            photo_file = d["Photo #2"].replace(" ","_").replace("'","")
            caption = d["Caption for Photo #2"]    
            image_table_text += an_image.format(photo_file = photo_file ,
                                           caption = caption )
            
        if type(d["Photo #3"]) == unicode:
            photo_file = d["Photo #3"].replace(" ","_").replace("'","")
            caption = d["Caption for photo #3"]    
            image_table_text += an_image.format(photo_file = photo_file ,
                                           caption = caption )
        image_table_text += "</div>"
        
        
    with open("Species_Image_HTML/"+d['Species file name']+"_images.html", 'w') as f:
        f.write(image_table_text)


    
```

```python
example_text = '''	<ul class="collapser">
					
					<li><a href="#"><i class="fas fa-plus-circle"></i> 1-26, Birds sing, and call</a>
						
						<ul class="show-hide">
							<li>1-5, <a href="recording.php">Black-capped chickadee</a></li>
							<li>6-11, <a href="recording.php">White-breasted nuthatch</a></li>
							<li>12-13, Veery</li>
							<li>14-21, American robin</li>
							<li>22-23, Gray catbird</li>
							<li>24-26, Common yellowthroat</li>
						</ul>
					</li>
					<li><a href="#"><i class="fas fa-plus-circle"></i> 27-41, Birds without a song just call</a>
					
						<ul class="show-hide">
							<li>27-29, Canada goose</li>
							<li>30-31, Mallard</li>
							<li>32, Chimney swift</li>
							<li>33-34, Killdeer</li>
							<li>35, Double-crested cormorant</li>
							<li>36-38, Red-shouldered hawk</li>
							<li>39, Red-bellied woodpecker</li>
							<li>40, American wigeon</li>
							<li>41, Pied-billed grebe</li>
						</ul>
						
					</li>
					<li><a href="#"><i class="fas fa-plus-circle"></i> 42-62, Female song, and duets</a></li>
					<li><a href="#"><i class="fas fa-plus-circle"></i> 63-78, Mechanical (non-vocal) sounds</a></li>
					<li><a href="#"><i class="fas fa-plus-circle"></i> 713-734, Extra! Extra! Bonus!--Australia</a></li>
					
				</ul>'''



recording_search_text = u'	<ul class="collapser">\n'


subchapter_heading_text = u'''					<li><a href="#"><i class="fas fa-plus-circle"></i>\
{subchapter_w_numbers}</a>
						
						<ul class="show-hide">
'''

species_text = u'''							<li>{numbers}, <a href="recording.php?page={page}">\
{species_name}</a></li>
'''

subchapter_footer = u'''						</ul>
						
					</li>'''


for subchapter in df['Simplified secondary chapter heading'].unique():
    if type(subchapter) != float:
        d_subchapter = df[df['Simplified secondary chapter heading'] == subchapter ]

        # get the recording range
        recordings = d_subchapter['Numbered sounds in file'].values
        recordings_ints = []
        for r in recordings:
            r = r.split("-")
            for rr in r:
                recordings_ints.append(int(rr))
        recordings_ints = np.array(recordings_ints)
        min_v = recordings_ints.min()
        max_v = recordings_ints.max()
        subchapter_heading = u"{}-{}, {}".format(min_v,max_v,subchapter)
        print(subchapter_heading)

        recording_search_text += subchapter_heading_text.format(subchapter_w_numbers=subchapter_heading)

        for index, row in d_subchapter.iterrows():
            numbers = row['Numbered sounds in file']
            species_name = row['Species name']
            filename = row['Species file name']
            value = df_pages[df_pages.SpeciesFile == filename]['Sequence of all items'].values[0]
    #         print(species_name, value)
            recording_search_text += species_text.format(page=value, numbers=numbers, species_name=species_name)

        recording_search_text += subchapter_footer

recording_search_text += u"				</ul>"  
    
with open("recording_lookup.php", 'w') as f:
    f.write(recording_search_text.encode("utf-8"))
```

```python
def get_species_index(speciesfile):
    value = df_pages[df_pages.SpeciesFile == speciesfile]['Sequence of all items'].values[0]
    return value
    
df['species_index'] = df['Species file name'].apply(get_species_index)
```

```python
def get_qr_url(speciesfile):
    value = df_pages[df_pages.SpeciesFile == speciesfile]['URL for QR codes'].values[0]
    return value
    
df['qr_url'] = df['Species file name'].apply(get_qr_url)
```

```python
def get_group_name(species_name):
    try:
        value = df_species[df_species.Species == species_name]['Group name'].values[0]
        return value
    except:
        return None

def get_family_name(species_name):
    try:
        value = df_species[df_species.Species == species_name]['Family name'].values[0]
        return value
    except:
        return None

df['group_name'] = df['Species name'].apply(get_group_name)

df['family_name'] = df['Species name'].apply(get_family_name)

```

```python
# "Group rank
# (alternative to column O)"	Group name	"Family rank
# (alternative to column Q)"	Family name	Species rank

def get_group_rank(species_name):
    try:
        value = df_species[df_species.Species == species_name]['''Group rank
(alternative to column O)'''].values[0]
        return value
    except:
        return None

def get_family_rank(species_name):
    try:
        value = df_species[df_species.Species == species_name]['''Family rank
(alternative to column Q)'''].values[0]
        return value
    except:
        return None

df['group_name_rank'] = df['Species name'].apply(get_group_rank)

df['family_name_rank'] = df['Species name'].apply(get_family_rank)
```
